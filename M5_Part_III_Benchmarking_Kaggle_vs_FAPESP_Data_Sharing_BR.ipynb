E) **ICU first two hours of hospitalization**: ALURA and Sirio-Libanes teams recomended that, when working with the Kaggle dataset, to deleted the patients on ICU, particularly the 32 patients who were recorded as in ICU during the first two hours from hospitalisation. 

It is curious to notice that in the e_mer_a_short - derived from HSL_Junho21 dataset in the column 'DE_ORIGEM', we have 51 patients for whom the request for covid-19 test was apparently done when the patient was already in ICU:
* 76% of them (39 out of 51) were >= 65 yo. 
* 76% of them (30 out of the 39 >= 65yo) were male.

That said, the Age_percentile distribution and the range of ages it includes, alerte us about a high number of subjects younger than 65 yo in the e_mer_a_short - derived from HSL_Junho21. However, as already stressed here above, an investigation is needed to understand why the number of younger than 65 yo is remarkbly higher than in the Kaggle dataset.
